## Implement Gradient Descent
y_predict=m1*weight+m2*engine_size+b

```python
m1=0.1
m2=0.7
b=0.5
iterations=10000
lrate=0.01
x1=data['weight'].values
x2=data['engine_size'].values
m=len(y)
x1 = (x1 - x1.mean()) / x1.std()
x2 = (x2 - x2.mean()) / x2.std()

lrate = 0.001
```

```python
for i in range(iterations):

    y_pred = m1 * x1 + m2 * x2 + b

    error = y_pred - y

    cost = (1/(2*m)) * np.sum(error**2)

    dm1 = (1/m) * np.sum(error * x1)
    dm2 = (1/m) * np.sum(error * x2)
    db  = (1/m) * np.sum(error)

    m1 = m1 - lrate * dm1
    m2 = m2 - lrate * dm2
    b  = b  - lrate * db

    if i % 1000 == 0:
        print("Iteration:", i, "Cost:", cost)
```
