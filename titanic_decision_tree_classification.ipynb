## download data

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("yasserh/titanic-dataset")

print("Path to dataset files:", path)
```

## import libraries

```python
import pandas as pd
import os 
from sklearn.preprocessing import OneHotEncoder
import warnings
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
warnings.filterwarnings('ignore')
```

## read data

```python
os.listdir(path)
```

```python
data=pd.read_csv(path+'/Titanic-Dataset.csv')
```

```python
data.head(5)
```

## data preprocessing

```python
data.shape
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data.Cabin.value_counts()
```

```python
data.Cabin.unique()
```

```python
data.drop('Cabin',axis=1,inplace=True)
```

```python
data.Embarked.value_counts()
```

```python
data[data['Embarked'].isna()]
```

```python
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
```

```python
data['Age'] = data['Age'].fillna(data.groupby('Pclass')['Age'].transform('mean'))
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
data.drop('Name',axis=1,inplace=True)
```

```python
data['Sex']=data['Sex'].map({'male':0,'female':1})
```

```python
data.Sex.value_counts()
```

```python
data.drop('Ticket',axis=1,inplace=True)
```

```python
data.Embarked.value_counts()
```

```python
encoder = OneHotEncoder(drop='first', sparse_output=False)

embarked_encoded = encoder.fit_transform(data[['Embarked']])
```

```python
embarked_df = pd.DataFrame(embarked_encoded, columns=encoder.get_feature_names_out(['Embarked']))

data = data.drop('Embarked', axis=1)
data = pd.concat([data, embarked_df], axis=1)
```

```python
data.info()
```

```python
data.Survived.value_counts()
```

## data splitting

```python
x=data.drop('Survived',axis=1)
y=data.Survived
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
accuracy_score(ytest,model.predict(xtest))
```

```python
precision_score(ytest,model.predict(xtest))
```

```python
recall_score(ytest,model.predict(xtest))
```

```python
f1_score(ytest,model.predict(xtest))
```

```python
cm=confusion_matrix(ytest,model.predict(xtest))
```

```python
plt.figure(figsize=(6,5))

sns.heatmap(cm, 
            annot=True, 
            fmt="d", 
            cmap="Greens",
            linewidths=0.5,
            linecolor="white")

plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.title("Confusion Matrix", fontsize=14, weight="bold")

plt.show()
```
